```python
import matplotlib.pyplot as plt 
from first_class import ComplexNumber, i
from math import sin, cos 
```

## Намалюємо спіраль Архімеда

```python
res = []
 
for i in range(1000): 
    res.append(i * ComplexNumber(sin(i), cos(i)))
        

# використання іграшкового метода. Працює довго.
# for num in res: 
#     num.draw()
# plt.show()
```

```python
def draw(seq, size=None, window=None): 
    plt.figure(figsize=(10, 10))
    real = [el.x for el in seq]
    imag = [el.y for el in seq]
    plt.scatter(real, imag, s=size)
    
    if window: 
        x, y = window
        plt.xlim([-x, x])
        plt.ylim([-y, y])
    
    plt.show()
    

draw(res)
```

## Тепер намалюємо сітку з комплексних чисел і спробуємо її якимось чином змінити.

```python
lattice = [
    ComplexNumber(i, j) for i in range(-40, 40) for j in range(-40, 40)
]

draw(lattice)
```

## Наприклад піднесемо всі числа до квадрату

```python
draw([z * z for z in lattice])
```

## Піднесемо до квадрату і домножимо на уявну одиницю

```python
draw([(z * z) * ComplexNumber(0, 1) for z in lattice])
```

## Візьмемо обернений до кожного 

```python
draw([z.inverse() for z in lattice if z.norm() != 0], size=0.4, window=[0.25, 0.25])
```

```python
draw([(z + 1) * (z - 1).inverse() for z in lattice if (z*z-1).norm() != 0], size=0.4)
```
